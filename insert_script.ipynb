```python
def insert_surah(nama, bagian=None):
    print("Start", end="> ")
    nama_surah = nama
    print(nama_surah)
    info_surah = info_surah_f(nama)
    if bagian == None:
        url = f"https://qurano.com/id/{nama_surah}/"
    else:
        url = f"https://qurano.com/id/{nama_surah}/{bagian}/"
        
    print("Load html", end=" ")
    html = asyncio.run(get_html(url))

    # Parsing HTML dengan BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
    # Lakukan parsing sesuai kebutuhan
    # print(soup.prettify())



    quran_container = soup.find_all(class_="quran-container")
    translations = soup.find_all(class_="translation")
    j_ayat = len(translations)
    print("Get and parse\nMenyimpan:", end=" ")

    for i in range(j_ayat):
        global langka
        langka += 1
        id_surah = info_surah[0]
        nama_surah = info_surah[1]
        id_ayat = langka #i+1
        text_ayat = translations[i].get_text()
        text_ayat_perkata = ""
        isiAyat = quran_container[i].find_all(class_="words-text")
        for i in range(len(isiAyat)):
            text_ayat_perkata += isiAyat[i].get_text() + " "
        try:
            surahlatest = Surah(id_surah=id_surah, nama_surah=nama_surah, id_ayat=id_ayat, text_ayat=text_ayat, text_ayat_perkata=text_ayat_perkata)
            db.session.add(surahlatest)
            db.session.commit()
            print(langka, end=" ")
        except:
            db.session.rollback()
            print("Gagal")
    del html
    del soup
    del quran_container
    del translations
    print("\nSelesai")
```

```python

from sqlalchemy.sql import text
from app.models import Surah
from app import app, db
app.app_context().push()

def initialize_fts():
    with db.engine.connect() as connection:
        # Membuat tabel FTS
        connection.execute(text('''
            CREATE VIRTUAL TABLE IF NOT EXISTS surah_fts USING fts5(
                id_surah, 
                nama_surah, 
                text_ayat, 
                id_ayat
            )
        '''))

        # Mengisi tabel FTS dengan data dari tabel Surah
        surahs = Surah.query.all()
        for surah in surahs:
            print("tes")
            connection.execute(text('''
                INSERT INTO surah_fts (id_surah, nama_surah, id_ayat, text_ayat)
                VALUES (:id_surah, :nama_surah, :id_ayat, :text_ayat)
            '''), {'id_surah': surah.id_surah, 'nama_surah': surah.nama_surah, 'id_ayat':surah.id_ayat, 'text_ayat': surah.text_ayat})
        connection.commit()


initialize_fts()
```
